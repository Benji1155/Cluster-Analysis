```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Print the data
data = pd.read_csv('DMV302_Assessment_2_AtRiskStudentsTraining.csv')

print(data)
```

```python
# Required libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_absolute_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
import joblib

# Load the training dataset
train_data = pd.read_csv('DMV302_Assessment_2_AtRiskStudentsTraining.csv')

# Preprocessing for training set
X_train = train_data[['GPA', 'attendance', 'duration', 'language']]  # Features
y_train = train_data['at-risk']                                       # Target

# Encode categorical variable
le = LabelEncoder()
X_train['language'] = le.fit_transform(X_train['language'])

# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# Implement KNN with K=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Save the model
joblib.dump(knn, 'knn_model.pkl')

# Load the test dataset
test_data = pd.read_csv('DMV302_Assessment_2_AtRiskStudentsTest.csv')

# Preprocessing for test set
X_test = test_data[['GPA', 'attendance', 'duration', 'language']]  # Features
y_test = test_data['at-risk']                                       # Target

# Encode categorical variabe
X_test['language'] = le.transform(X_test['language'])

# Feature Scaling for test set
X_test = scaler.transform(X_test)

# Make predictions on the test set and evaluating the model
y_test_pred = knn.predict(X_test)

# Calculate accuracy and error
test_accuracy = accuracy_score(y_test, y_test_pred)
mae = mean_absolute_error(y_test, y_test_pred)
confusion = confusion_matrix(y_test, y_test_pred)
report = classification_report(y_test, y_test_pred)

# Print test results
print(f"Test Accuracy: {test_accuracy:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print("Confusion Matrix:")
print(confusion)
print("Classification Report:")
print(report)

# Save test results to a file 
with open("KNN_test_results.txt", "w") as f:
    f.write(f"Test Accuracy: {test_accuracy:.2f}\n")
    f.write(f"Mean Absolute Error: {mae:.2f}\n")
    f.write("Confusion Matrix:\n")
    f.write(str(confusion) + "\n")
    f.write("Classification Report:\n")
    f.write(report)

# Vary the number of neighbors and measuring the error
neighbor_values = [1, 2, 3, 5, 7, 9]  # Testing different values of k
errors = []

for k in neighbor_values:
    # Create a new KNN model for each value of k
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_test_pred = knn.predict(X_test)
    
    # Calculate error using MAE
    mae = mean_absolute_error(y_test, y_test_pred)
    errors.append((k, mae))
    
    # Print the results for each k
    print(f"K={k}, Mean Absolute Error (MAE): {mae:.2f}")

# Display the results
print("\nErrors for different values of K:")
print("Steps   n_neighbors   Error (MAE)")
for k, error in errors:
    print(f"Step    {k}            {error:.2f}")

# Save the table to a file
with open("KNN_neighbor_errors.txt", "w") as f:
    f.write("Steps   n_neighbors   Error (MAE)\n")
    for k, error in errors:
        f.write(f"Step    {k}            {error:.2f}\n")

```
